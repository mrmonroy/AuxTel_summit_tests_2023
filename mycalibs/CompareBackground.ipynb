```python
DATE1="20230926"
DATE2="20230914"
files = ['background_20230926_empty~holo4_003_506.csv','background_20230914_collimator~holo4_003_158.csv',
'background_20230914_empty~holo4_001_235.csv',
'background_20230914_empty~holo4_003_238.csv']
```

```python
filterdispersers = ['empty~holo4_003','collimator~holo4_003','empty~holo4_001','empty~holo4_003']
seqnums = [506,158,235,238]
```

```python
N = len(files)
```

```python
all_df = []
for idx in range(N):
    df = pd.read_csv(os.path.join(path,files[idx]))
    all_df.append(df)
```

```python
mapampid = [0,1,2,3,4,5,6,7,15,14,13,12,11,10,9,8]
mapname = ['C10' , 'C11' ,'C12', 'C13', 'C14', 'C15', 'C16', 'C17', 'C07', 'C06', 'C05', 'C04', 'C03', 'C02', 'C01', 'C00'] 
```

```python
idxf=0
filterdisperser= filterdispersers[idxf]
seqnum = seqnums[idxf]
```

```python
fig, axs = plt.subplots(2, 8,figsize=(20,6),sharey=True)
idx=0
for ax in axs.ravel():
    name = mapname[idx]
 
    data = all_df[idxf][name]
    data0 = all_df[1][name]
    data2 = all_df[3][name]
    data1 = all_df[2][name]
    ax.hist(data,bins=50,range=(-100,100),facecolor="yellow")
    ax.hist(data0,bins=50,range=(-100,100),facecolor="b")
    ax.hist(data,bins=50,range=(-100,100),facecolor="red")
    ax.hist(data2,bins=50,range=(-100,100),histtype='step',color="r",lw=2)
    ax.hist(data1,bins=50,range=(-100,100),histtype='step',color="g",lw=2)
    x = data
    mu = data.mean()
    median = np.median(data)
    sigma = x.std()
    textstr = '\n'.join((
    r'$\mu=%.1f$' % (mu, ),
    r'$\mathrm{med}=%.1f$' % (median, ),
    r'$\sigma=%.1f$' % (sigma, )))

    ax.set_title(name)  
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.2)

    # place a text box in upper left in axes coords
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10,verticalalignment='top', bbox=props)
    print(f"{idx} ,{mapampid[idx]}, {name} ")
    idx+=1
    
the_title = f"{DATE1} : {filterdispersers[0]}(full red,{seqnums[0]}), {DATE2} {filterdispersers[1]}(full blue,{seqnums[1]}) {filterdispersers[3]}(empty red,{seqnums[3]}), {filterdispersers[2]}(empty green,{seqnums[2]}) "
plt.suptitle(the_title,fontsize=18)
plt.tight_layout()
```
